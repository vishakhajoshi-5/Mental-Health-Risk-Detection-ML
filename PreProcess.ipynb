```python
import pandas as pd
import os
```

```python
# Load the datasets

d1_original = pd.read_csv("datasets/dreaddit-train.csv")
d2_original = pd.read_csv("datasets/reddit_train.csv")
d3_original = pd.read_csv("datasets/train.csv")
```

```python
# Create a copy of the datasets

d1_original.to_csv("copy/d1.csv", index=False)
d2_original.to_csv("copy/d2.csv", index=False)
d3_original.to_csv("copy/d3.csv", index=False)
```

```python
# FOR DATASET D1
```

```python
d1 = pd.read_csv("copy/d1.csv")
d1.head(3)
```

```python
# CLEAN THE DATASET D1
```

```python
# Remove Duplicates
d1.drop_duplicates(inplace=True)

# Remove Post word:
d1["post"] = d1["post"].str.replace(r"(?i)^post:\s*", "", regex=True)

# Remove Question word:
d1["question"] = d1["question"].str.replace(r"(?i)Question:\s*", "", regex=True)

# Extract Yes/No
d1["stress_label"] = d1["response"].str.extract(r"(?i)^(yes|no)")

# Extract Reasoning Part
d1["stress_label"] = d1["response"].str.extract(r"(?i)^(yes|no)")


# Extract Reasoning Part
d1["reasoning"] = d1["response"].str.split("Reasoning:", expand=True)[1]
# Clean reasoning column
d1["reasoning"] = d1["reasoning"].str.strip()


# Lowercase Everything
d1["post"] = d1["post"].str.lower()
d1["question"] = d1["question"].str.lower()
d1["reasoning"] = d1["reasoning"].str.lower()
d1["stress_label"] = d1["stress_label"].str.lower()


# Remove NaN values 
d1 = d1.dropna()

# Drop Old Response Column
d1.drop(columns=["response"], inplace=True)


d1.head(3)
```

```python
# Label Encode Yes/No

d1["stress_label"] = d1["stress_label"].map({"yes": 1, "no": 0})
d1["stress_label"].value_counts()

d1.head(3)
```

```python
# Safe back the file into the d1

d1.to_csv("copy/d1.csv", index=False)
```
