```python
d1 = pd.read_csv("copy/d1.csv")
d1.head(3)
```

```python
# Remove Post:
d1["post"] = d1["post"].str.replace(r"(?i)^post:\s*", "", regex=True)

# Extract label from response (remove reasoning)
d1["label"] = d1["response"].str.split("Reasoning:", expand=True)[0]

# Clean label
d1["label"] = d1["label"].str.strip().str.replace(".", "", regex=False)

# Lowercase
d1["post"] = d1["post"].str.lower()
d1["label"] = d1["label"].str.lower()

# Remove duplicates
d1.drop_duplicates(inplace=True)

d1.head(3)
```
